In this example I will be attempting to import data from a csv file

Import the necessary modules

```python
import numpy as np
import pandas as pd
```

Store the file name in a string that you want to import data from

```python
filename = 'sitka_weather_2014.csv'

# Read the csv file and store it as a pandas Data Frame
df = pd.read_csv(filename)

# Output the object type to see if it was created successfully
type(df)
```

Let's attempt to view the first few rows of the data from the CSV file:

```python
df.head()
```

Lets Output the entire Data Frame now to see out entire CSV file contents:

```python
df
```

Yikes, this is a lot of data, 365 cases (data from each day of the year)

We can use the .loc function to specify which columns of data we wish to see

```python
# Return all observations of MaxTemperatureF
df.loc[:, "Max TemperatureF"]
```

We were able to return all the data from the Max Temperature variable. Using loc(), the first argument indicates the row and the second indicates the colum. In this case we used a : to return all rows from the table and we specified the column by using a string that matches the name of the column. You can place the names of multiple columns as a Python list instead of just using one column name. This will return more than one column. 

```python
# Return the data from the first 3 rows and the Max Temperature, Min Temperature, as well the the Event that day
df.loc[:3,["Max TemperatureF", "Mean TemperatureF", "Min TemperatureF"]]
```

This line only works for data that is NOT missing. If a variable is has a missing entry, than pandas will not allow it to work properly. 

Next, I will use .iloc() function to return data entries. .iloc() is an integer based slice
In the next example, I will return the first 3 rows of data from the file.

```python
df.iloc[:3]
```

This returned the first 3 rows of data with all the variables.

.iloc() can also return information from specific columns as well in addition to specific rows.

```python
df.iloc[353:, 1:4]
```

This returned information from starting from the row 353 to the end to the end of the file. We also returned information from the second, third, and fourth rows as well.
